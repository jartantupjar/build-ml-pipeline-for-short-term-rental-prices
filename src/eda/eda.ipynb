## Exploratory Data Analysis on the NYC AIRBNB DATASET

```python
import wandb
import pandas as pd
import pandas_profiling
```

###  load in the airbnb dataset

```python
#intialize w&b
run = wandb.init(project="nyc_airbnb", group="eda", save_code=True)
#download file from w&b
local_path = wandb.use_artifact("sample.csv:latest").file()
#load data 
df = pd.read_csv(local_path)
```

## explore the data

```python
profile = pandas_profiling.ProfileReport(df)
profile.to_widgets()
```

```python
# test for additional context in name column
display('beach instance',df[df['name'].str.contains("beach", na=False)].head(10))
display('luxury instance',df[df['name'].str.contains("luxury", na=False)].head(10))
display('sq instance',df[df['name'].str.contains("sq|sqft", na=False)].head(10))
```

```python
#explore number of reviews more
df[df['number_of_reviews']==0]
```

```python
#explore minimum nights for outliers 
df['minimum_nights'].value_counts()[:20]
```

### Findings:
 - price has quite a few outliers that makes its distribution very odd.
 - There are listings that are "Free" or zero. This must be a wrong entry as integers/floats don't usually allow nulls and mustve been replaced with zeroes. The upper limit however may need to be tweaked. There are instances where price exceeds 1000 but does skew our dataset by quite a bit. This value may need to be tweaked/adjusted accordingly
 - there are a lot of listings which has 0 number_of_reviews(10052 rows). This may need to be explored further during training phase as this column may negatively affect our models bias.
 -  longitude, latitude and price has correlation which makes sense from a logical standpoint
 -  minimum_nights appears to have some potential outliers. There are plenty of locations which require more than >90 days and one location in particular requires 1250 day requirement.
 - the column "name" appears to have additional contextual information that may tie/help explain their listing price further which could be extracted. Eg. there are places that are listed as apartments/house. Some locations mention they are close to the beach. some locations mention how big the apartment is.


### simple outlier evaluation and removal
 - try different values (this is not final)
 - based on an educated guess anything lower than 10  and anything higher than at least 350 should be too extreme. the higher bound can be tweaked/experimented with.

```python
# Drop outliers
min_price = 10
max_price = 350
idx = df['price'].between(min_price, max_price)
df = df[idx].copy()
# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'])
```
